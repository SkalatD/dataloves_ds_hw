3. Задана матриця `D` розмірністю 2х2. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на матрицю `D` і який результат буде. А потім виконайте множення і виведіть результат на екран.

```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
# Транспонуємо матриці обидві а тоді вже транспоновані помноживо, щоб отримати 2х2
first_task_matrix = np.dot(A.T, B.T)
print(f"Перше завдання: \n{first_task_matrix}\n")
# В другому завданнi я думаю, що отримаю матрицю (вектор) 2x1
second_task_vector = np.dot(first_task_matrix, C)
print(f"Друге завдання: \n{second_task_vector}\n Розмірність: {second_task_vector.shape}\n")
# В третьому завданнi ми множимо матрицю 2х2 на матрицю (вектор) 2х1, множити можна і отримаємо матрицю (вектор) 2х1
third_task_vector = np.dot(D, second_task_vector)
print(f"Третє завдання: \n{third_task_vector}\n Розмірність: {third_task_vector.shape}\n")
# Але, якщо в третьому завданнi множити навпаки вектор 2х1 на матрицю 2х2, то множити не можна, бо к-сть стовпців вектора не дорівнює к-сті рядків матриці
try:
  third_task_vector_error = np.dot(second_task_vector, D)

except ValueError as error:
  print(f"О ні, ми отримали помилку:\n {error}")
```
