### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Задаю символічну змінну
x = sp.symbols("x")
# Задаю функцію залежності оцінки від кількості годин, витрачених на навчання
f_learn = 10 * x - 5
# Знаходжу похідну для функції залежності оцінки від кількості годин, витрачених на навчання
derivative_f_learn = sp.diff(f_learn, x)

print(f"Похідна функції залежності оцінки від кількості годин, витрачених на навчання: {derivative_f_learn}")
```

```python
# Будую графіки функції та її похідної в діапазоні значень від 0 до 10 годин

# Діапазон значень x
x = np.linspace(0, 10, 100)  # від 0 до 10 годин, 100 точок

f_learn = 10 * x - 5
df_learn = [10]*100  # Похідна (вона є сталим числом, тому формую список зі 100 сталих чисел 10)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f_learn, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df_learn, label="f'(x) = 10", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції 10x - 5 та її похідної 10")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

*Оскільки похідна - це швидкість зміни функції в кожній точці, то виходить, що з плином часу (на 1 збільшення) додається 10 балів. Тому тим більше вчитися - тим більше отримаємо балів. Якщо 0 годин, то -5 балів, якщо 1 годину, то 5 балів, якщо 2 години, то 15 балів, якщо 2.5 години, то 20 балів.*

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# Задаю символічну змінну
x = sp.symbols("x")
# Задаю функції
f_sin = sp.sin(x)
f_ln = sp.log(x)
# Знаходжу похідні для функцій
derivative_f_sin = sp.diff(f_sin, x)
derivative_f_ln = sp.diff(f_ln, x)

print(f"Похідна функції sin(x): {derivative_f_sin}")
print(f"Похідна функції ln(x): {derivative_f_ln}")
```

```python
# Будую графік функції sin(x) та її похідної

# Діапазон значень x
x = np.linspace(-10, 10, 600)  # від -10 до 10, 600 точок

f_sin = np.sin(x)  # Функція
df_sin = np.cos(x)  # Похідна

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f_sin, label='f(x) = sin(x)', color='blue')
plt.plot(x, df_sin, label="f'(x) = cos(x)", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції sin(x) та її похідної cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Будую графік функції ln(x) та її похідної

# Діапазон значень x
x = np.linspace(1, 20, 100)  # від 1 до 20, 100 точок

f_ln = np.log(x)  # Функція
df_ln = 1/x  # Похідна

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f_ln, label='f(x) = ln(x)', color='blue')
plt.plot(x, df_ln, label="f'(x) = 1/x", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції ln(x) та її похідної 1/x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Задаю символічну змінну
x = sp.symbols("x")  # кількість вироблених одиниць
# Задаю функцію залежності загальних витрат компанії від обсягу виробництва
f_costs = x ** 2 - 5 * x + 1
# Знаходжу похідну для функції залежності загальних витрат компанії від обсягу виробництва
derivative_f_costs = sp.diff(f_costs, x)

print(f"Похідна функції залежності загальних витрат компанії від обсягу виробництва: {derivative_f_costs}")
```

```python
# Будую графік функції x² - 5x + 1 та її похідної

# Діапазон значень x
x = np.linspace(0, 100, 50)  # від 0 до 100, 500 точок

f_costs = x ** 2 - 5 * x + 1  # Функція
df_costs = 2*x - 5  # Похідна

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f_costs, label='f(x) = x² - 5x + 1', color='blue')
plt.plot(x, df_costs, label="f'(x) = 2x - 5", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції x² - 5x + 1 та її похідної 2x - 5")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
